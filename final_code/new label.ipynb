```python
import pandas as pd
import json
import os

## 필사본 train 폴더
path = "/home/piai/바탕화면/hanja_crop_50/new50/output_crop/train"
folder_list = os.listdir(path)
len(folder_list)

# 데이터 로드
label_df = pd.read_csv('/home/piai/바탕화면/hanja_crop_50/new50/haeseo_label.csv', names = ['label', 'idx'], header=None)
label_df.head()
```

```python
## 원본 label에서 바뀐 index 순서 저장
i = 0
for no, row in label_df.iterrows():
    if str(row['idx']) in folder_list:  
        label_df.at[no, 'new_idx'] = int(i)
        i+=1
```

```python
label_df.dropna(subset=['new_idx']).reset_index(drop = True)
new_df = label_df.dropna(subset=['new_idx']).reset_index(drop = True)
new_df = new_df.astype({"new_idx":'int'})
new_df
```

```python
## 바뀐 index 파일 저장
new_df.to_csv("/home/piai/바탕화면/hanja_crop_50/new50/new50_label.csv", index=False)
```
